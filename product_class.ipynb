```python
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
df = pd.read_csv("data/foiclass.txt", sep="|", encoding='ISO-8859-1' )
df.sample(5)
```

```python
df.info()
```

```python
df["UNCLASSIFIED_REASON"] = df["UNCLASSIFIED_REASON"].astype(str)
df["REVIEWCODE"] = df["REVIEWCODE"].astype(str)
df["REGULATIONNUMBER"] = df["REGULATIONNUMBER"].astype(str)
df["SUBMISSION_TYPE_ID"] = df["SUBMISSION_TYPE_ID"].astype(str)
```

```python
prof = ProfileReport(df)
prof
```

```python
prof.to_file(output_file='reports/profile.html')
```
